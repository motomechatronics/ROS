## 1. How to navigate through a Linux filesystem

Use _pwd_ to show position between folders. An example of using _pwd_ command is shown below:

```python
pwd
```

Use _cd_ to navigate through the folders. Below is an example of using _cd_:

```python
cd /home/jovyan/
```

To go to a previous folder, for example from _jovian_ to _home_, type:

```python
cd ..
```

View the contents of a directory by typing:

```python
ls
```

To show hidden files

```python
ls -la
```

## 2. How to interact with Linux filesystem

Create a directory typing:

```python
mkdir example_directory
```

You can create a file inside the _example_directory_ folder by _touch_ command:

```python
cd example_directory
```

```python
touch myfile
```

## 3. How to edit files using the shell

In Ubuntu, there are several text editors such as _nano_, _gedit_ or _vi_. You can use the latest to edit previous my_file:

```python
vi myfile
```

Inside _vi_, you can:
- to type the _i_ letter to insert the text;
- press _esc_ botton to exit from insert mode;
- to type _:w_ to save;
- to type _:q_ to exit from _vi_;
- to type _:wq_ to save and exit.

## 4. Manage access to files

Finally, you can give _myfile_ execute permission using the _chmod_ command as follows. Often it is necessary to use this command in conjunction with _sudo_.

```python
sudo chmod +x myfile
```
